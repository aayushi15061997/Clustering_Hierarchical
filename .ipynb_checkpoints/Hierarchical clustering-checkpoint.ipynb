```python
#There are two types of hierarchical clustering
#1-agglomerative(Bottoms up approach)
#2-divisive (starting from the top and dividing as we go down)
```

```python
#steps for Agglomerative HC
#1-Make each data point a single point cluster i.e. N -clusters
#2-take the two closest data points and make them as one cluster (N-1 clusters now at this step)
#2-take the two closest Clusters and make them as one cluster (N-2 clusters )
#3-repeat the step 3 until only one cluster is left
#distance btw clusters can be found by using-
#1-closest points          #2-farthest point
#3-average distance        #4-dist. btw centroids
```

```python
#dendograms- MEMORY of HC algo ,it remembers every step that we did
#connect clusters with a vertical line with y axis as euclidean dist
#between them(dissimilarity between them)
#find the largest vertical dist that doesn't cross a horizontal line
#and set a dist threshold so we can extract the no. of optimal clus
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values

#using the dendogram to find the optimal number of clusters(n_clusters for the next step)
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(x ,method = 'ward'))
#ward method is used to minimise the within cluster variance
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()

#Fitting HC to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters =5,affinity ='euclidean',linkage ='ward')
y_hc = hc.fit_predict(x)

#visualising the clusters in 2-D
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='red',label='Careful')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='blue',label='Standard')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,c='green',label='Target')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,c='cyan',label='Careless')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,c='magenta',label='Sensible')

plt.title('Clusters of Clients')
plt.xlabel('Annual Income(k$)')
plt.ylabel('spending score(1-100)')
plt.legend()
plt.show()  
```
